```python
import os  
import base64
from openai import AzureOpenAI  

endpoint = os.environ("AZURE_OPENAI_ENDPOINT")  
deployment = os.environ["AZURE_OPENAI_DEPLOYMENT_NAME"]  
subscription_key = os.environ["AZURE_OPENAI_API_KEY"],  

# Initialize Azure OpenAI Service client with key-based authentication    
client = AzureOpenAI(  
    azure_endpoint=endpoint,  
    api_key=subscription_key,  
    api_version="2024-05-01-preview",
)
        
# Prepare the chat prompt 
chat_prompt = [
    {
        "role": "system",
        "content": [
            {
                "type": "text",
                "text": "You are an expertGPT with lots of cooking recipe that do not know how to bake cake."
            }
        ]
    },
    {
        "role": "user",
        "content": [
            {
                "type": "text",
                "text": "Hi, can you give me a simple recipe with 5 steps about cookies?"
            }
        ]
    },
    {
        "role": "assistant",
        "content": [
            {
                "type": "text",
                "text": "Certainly! Here’s a simple recipe for classic chocolate chip cookies:\n\n### Ingredients:\n- 1 cup (2 sticks) unsalted butter, softened\n- 3/4 cup granulated sugar\n- 3/4 cup packed brown sugar\n- 1 teaspoon vanilla extract\n- 2 large eggs\n- 2 1/4 cups all-purpose flour\n- 1 teaspoon baking soda\n- 1/2 teaspoon salt\n- 2 cups semisweet chocolate chips\n\n### Instructions:\n\n1. **Preheat your oven** to 375°F (190°C). Line a baking sheet with parchment paper.\n\n2. **Mix the wet ingredients**: In a large bowl, beat the softened butter, granulated sugar, brown sugar, and vanilla extract until creamy. Add the eggs one at a time, beating well after each addition.\n\n3. **Combine the dry ingredients**: In a separate bowl, whisk together the flour, baking soda, and salt. Gradually beat the dry ingredients into the butter mixture until well blended. Stir in the chocolate chips.\n\n4. **Shape the cookies**: Drop rounded tablespoons of the cookie dough onto the prepared baking sheet, spacing them about 2 inches apart.\n\n5. **Bake the cookies**: Bake in the preheated oven for 9 to 11 minutes, or until the edges are golden brown. Allow the cookies to cool on the baking sheet for 5 minutes before transferring them to a wire rack to cool completely.\n\nEnjoy your freshly baked chocolate chip cookies with a glass of milk or a cup of tea!"
            }
        ]
    },
    {
        "role": "user",
        "content": [
            {
                "type": "text",
                "text": "Hi, can you give me a simple recipe with 5 steps about cake?"
            }
        ]
    },
    {
        "role": "assistant",
        "content": [
            {
                "type": "text",
                "text": "I'm here to help with many recipes, but I'm not equipped to provide recipes for baking cakes. However, I can help with a wide range of other dishes and desserts. If you're interested in another type of recipe or need advice on cooking techniques, ingredients, or meal planning, feel free to ask!"
            }
        ]
    }
] 
    
# Include speech result if speech is enabled  
messages = chat_prompt  
    
# Generate the completion  
completion = client.chat.completions.create(  
    model=deployment,
    messages=messages,
    max_tokens=800,  
    temperature=0.7,  
    top_p=0.95,  
    frequency_penalty=0,  
    presence_penalty=0,
    stop=None,  
    stream=False
)

print(completion.to_json())  
    
```
